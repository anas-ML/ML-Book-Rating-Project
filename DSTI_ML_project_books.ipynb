```python
#transform "publication_date" column into datetimetime data type. However, when executing the transformation, an error message
#is returned as some string values do not have relevant date format (%m%d%Y).Hence, i use the "try - except" method to identify
#the values with inconsistent formatting

from datetime import datetime
def valid_date_string(string_x):
    if not isinstance(string_x,str):
        string_x = str(string_x)
    string_x = string_x.strip()
    if len(string_x.split("/")) != 3:
        return False
    try:
        datetime.strptime(string_x,"%m/%d/%Y")
        return True
    except ValueError:
        return False
```

```python
#identiy invalid date formats

invalid_dates = df.loc[df["publication_date"].apply(valid_date_string)==False,:]
invalid_dates
```

```python
#web search shows that there are publication dates for these 2 books that very close to wrong publication dates values mentionned in the 
#dataset for these 2 books: 10/31/2000 for the first book, 6/1/1982 for the second. This evidences that there were data input errors in the initial
#dataset. I will hence replace these 2 dates by the new values.

index = invalid_dates.index
df.iloc[index,9] = ["10/31/2000","6/1/1982"]
```

```python
#verify the transformation
df.iloc[index]
```

```python
#now we can transform the column "publication_date" to pandas datetime data type. The function normalize() maintains only the
#and removes time entries.
df["publication_date"] = pd.to_datetime(df["publication_date"],format="%m/%d/%Y").dt.normalize()
df.dtypes
```

```python
#now i start the transformation of the isbn columns. Values are stored as "object", with notations such as "9.78044E+12", along with
#alphanumeric values, such as "081221627X". i would like to store all the values as alphanumeric strings.
#at this stage, i don't know yet if these columns may bring any valuable information to our problem. I keep the columns for now, these may
#be removed at a later stage of the project.
#first we define a function to verify if a value has a numeric format.

def valid_numeric(x):
    try:
        pd.to_numeric(x)
        return True
    except ValueError:
        return False    
```

```python
#test the function for specific cases
valid_numeric("9.78044E+12")
```

```python
#test the function for specific cases
valid_numeric("081221627X")
```

```python
#test the function for specific cases
valid_numeric(9.78044E+12)
```

```python
#define a function to process isbn columns values.

def process_isbn(x):
    if valid_numeric(x):
        x = int(pd.to_numeric(x))
        x = str(x)
        x = x.strip()
        return x
    else:
        x = str(x)
        x = x.strip()
        return x
```

```python
#test the function for specific cases
process_isbn("9.78044E+12")
```

```python
#test the function for specific cases
process_isbn("081221627X")
```

```python
#test the function for specific cases
process_isbn(9.78044E+12)
```

```python
#apply the transformation to "isbn" column
df["isbn"] = df["isbn"].apply(process_isbn)
```

```python
#apply the transformation to "isbn13" column
df["isbn13"] = df["isbn13"].apply(process_isbn)
```

<h3>Author and publisher average rating_(on train data)</h3>
here, i will further enhance the feature engineering phase. initially, i avoided using the target values for the transformation of the different
features. however, at this stage and given the limitations within the datased, i made the choice to replace author and publisher features with 
their corresponding average rating and ratings_count. we can ensure to avoid any data leakage between training and test data by applying this 
transformation on the training subset only.
in anycase, in a real life scenario, at prediction time, we can have access to the historic athor/publisher's statistics. Hence, for the testing 
subset, we will assign the statistics from the training subset, and in case the book does not exist in the training subset (equivalent to a new
author/publisher in a real life scenario), we will assign the overall average rating from the training subset.

```python
#get a previous version of df to recover author/publisher values 
df = df_4_3
```

```python
df.columns
```

```python
#drop unwanted columns
columns = ['title','isbn','isbn13','language_code','text_reviews_count']
df = df.drop(columns=columns)
```

```python
#extract publication year feature (let's test if it may have impact on the model's performance)
df["publication_year"] = df["publication_date"].apply(lambda x: x.year)
df = df.drop(columns="publication_date")
```

```python
df.columns
```

```python
#split the dataset
df_train, df_test = train_test_split(df, test_size=0.2, random_state=20)
```

```python
#reproduce author_stats and publisher_stats and merge these to the books dataframe

author_stats = df_train.groupby("author_1").agg(
    average_author_rating=("average_rating", "mean"),
    average_author_rating_count=("ratings_count_log", "mean")
).reset_index()

publisher_stats = df_train.groupby("publisher").agg(
    average_publisher_rating=("average_rating", "mean"),
    average_publisher_rating_count=("ratings_count_log", "mean")
).reset_index()

df_train = df_train.merge(author_stats, on="author_1", how="left")
df_train = df_train.merge(publisher_stats, on="publisher", how="left")
```

```python
df_train.columns
```

```python
#add new columns to the tst subset to align it with the training suset. prepare functions and transform the test subset 
#as per the intended approach

df_test["average_author_rating"] = None
df_test["average_author_rating_count"] = None
df_test["average_publisher_rating"] = None
df_test["average_publisher_rating_count"] = None

def process_average_author_rating(x):
    if x in(df_train["author_1"]).values:
        x = df_train[(df_train["author_1"]==x)]["average_author_rating"].iloc[0]
        return x
    else: 
        x = df_train["average_rating"].mean()
        return x

def process_average_author_rating_count(x):
    if x in(df_train["author_1"]).values:
        x = df_train[(df_train["author_1"]==x)]["average_author_rating_count"].iloc[0]
        return x
    else: 
        x = df_train["ratings_count_log"].mean()
        return x

def process_average_publisher_rating(x):
    if x in(df_train["author_1"]).values:
        x = df_train[(df_train["author_1"]==x)]["average_publisher_rating"].iloc[0]
        return x
    else: 
        x = df_train["average_rating"].mean()
        return x

def process_average_publisher_rating_count(x):
    if x in(df_train["author_1"]).values:
        x = df_train[(df_train["author_1"]==x)]["average_publisher_rating_count"].iloc[0]
        return x
    else: 
        x = df_train["ratings_count_log"].mean()
        return x

df_test["average_author_rating"] = df_test["author_1"].apply(process_average_author_rating)
df_test["average_author_rating_count"] = df_test["author_1"].apply(process_average_author_rating_count)
df_test["average_publisher_rating"] = df_test["author_1"].apply(process_average_publisher_rating)
df_test["average_publisher_rating_count"] = df_test["author_1"].apply(process_average_publisher_rating_count)
```
